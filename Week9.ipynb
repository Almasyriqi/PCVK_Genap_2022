> **Dilasi Tanpa Library** 

3. Buatlah fungsi dilasi serta tampilkan hasil dilasi pada citra dengan Structuring Element 
Dilasi (SED) berbentuk square 5 x 5 tanpa menggunakan library morphology dari 
openCV (Gunakan gambar j.png)

```python
# Dilasi tanpa library

# Input image
input_image = cv2.imread('/content/drive/MyDrive/PCVK_MM/Images/j.png')
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)

# Membuat structured element berbentuk square 5x5
se = np.ones((5,5))

def morph_dilate(img, kernel):
    # Menentukan baris dan kolom citra
    rows = img.shape[0]
    cols = img.shape[1]
    dilate_img=np.zeros((rows,cols))

    newimg=np.zeros((img.shape[0]+2,img.shape[1]+2))
    newimg[1:-1,1:-1]=img
    for i in range(0,rows-(kernel.shape[0]-1)):
        for j in range(0,cols-kernel.shape[0]-1):
            dilate_img[i,j]=np.max(se * newimg[i:i+kernel.shape[0],j:j+kernel.shape[0]])

    return dilate_img

# Citra hasil dilasi
output_image = morph_dilate(input_image, se)

# Membuat figure
fig, axs = plt.subplots(1, 2, figsize=(9,5))

# Menampilkan citra awal
axs[0].set_title("Citra Awal", fontSize=18)
axs[0].imshow(input_image, cmap='gray')

# Menampilkan citra hasil dilasi
axs[1].set_title("Citra Hasil Dilasi", fontSize=18)
axs[1].imshow(output_image, cmap='gray')
```

>**Erosi Tanpa Library** 

4. Buatlah fungsi erosi serta tampilkan hasil erosi pada citra dengan Structuring Element 
Erosi (SE) berbentuk square 5 x 5 tanpa menggunakan library morphology dari 
openCV (Gunakan gambar j.png)

```python
# Erosi tanpa library

# Input image
input_image = cv2.imread('/content/drive/MyDrive/PCVK_MM/Images/j.png')
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)

# Membuat structured element berbentuk square 5x5
se = np.ones((5,5))

def morph_erode(img, kernel):
    # Menentukan baris dan kolom citra
    rows = img.shape[0]
    cols = img.shape[1]
    erode_img=np.zeros((rows,cols))

    newimg=np.zeros((img.shape[0]+2,img.shape[1]+2))
    newimg[1:-1,1:-1]=img
    for i in range(0,rows-(kernel.shape[0]-1)):
        for j in range(0,cols-(kernel.shape[0]-1)):
            erode_img[i,j]=np.min(se * newimg[i:i+kernel.shape[0],j:j+kernel.shape[0]])
    return erode_img

# Citra hasil erosi
output_image = morph_erode(input_image, se)

# Membuat figure
fig, axs = plt.subplots(1, 2, figsize=(9,5))

# Menampilkan citra awal
axs[0].set_title("Citra Awal", fontSize=18)
axs[0].imshow(input_image, cmap='gray')

# Menampilkan citra hasil erosi
axs[1].set_title("Citra Hasil Erosi", fontSize=18)
axs[1].imshow(output_image, cmap='gray')
```

> **Operasi Opening**

7. Buatlah program implementasi operasi opening dengan Structuring Element 
berbentuk square 7 x 7 tanpa dan dengan menggunakan library morphology dari 
openCV (Gunakan gambar j.png)

```python
# Implementasi operasi opening

# Input image
input_image = cv2.imread('/content/drive/MyDrive/PCVK_MM/Images/j.png')
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)

# Membuat structured element berbentuk square 7x7
se = np.ones((7,7))

# Hasil operasi opening tanpa menggunakan library
opening_image = morph_dilate(morph_erode(input_image, se), se)

# Hasil operasi opening menggunakan library opencv
output_image = cv2.morphologyEx(input_image, cv2.MORPH_OPEN, se)

# Membuat figure untuk hasil operasi tanpa library
fig, axs = plt.subplots(1, 2, figsize=(9,5))
fig.suptitle('Operasi Opening Tanpa Library', fontsize=18)
# Menampilkan citra awal
axs[0].set_title("Citra Awal", fontSize=16)
axs[0].imshow(input_image, cmap='gray')

# Menampilkan citra hasil opening tanpa library
axs[1].set_title("Citra Hasil Opening", fontSize=16)
axs[1].imshow(opening_image, cmap='gray')

# Membuat figure untuk hasil operasi opening dengan library
fig1, axs1 = plt.subplots(1, 2, figsize=(9,5))
fig1.suptitle('Operasi Opening Dengan Library', fontsize=18)
# Menampilkan citra awal
axs1[0].set_title("Citra Awal", fontSize=16)
axs1[0].imshow(input_image, cmap='gray')

# Menampilkan citra hasil operasi opening dengan library
axs1[1].set_title("Citra Hasil Opening", fontSize=16)
axs1[1].imshow(output_image, cmap='gray')
```
