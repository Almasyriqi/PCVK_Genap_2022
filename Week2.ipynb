Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
# pembuatan black image dengan tipe data int32.
black_img2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

```python
# Berikut adalah kode program untuk inisialisasi NumPy array dengan tpe data int32
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
```

```python
# Array tersebut kemudian di reshape sebagai berikut
pts = vertices.reshape((-1,1,2)) # nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yang mewakili R, G, B
pts
```

```python
# Penambahan polyline pada black image kedua yang telah dibuat.
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```

```python
# Tugas 1
# Menampilkan image dalam channel Red-Green 
img6 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img6[:, :, 2] = 0

# Menampilkan image dalam channel Green-Blue 
img7 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img7[:, :, 0] = 0

finalFrame = cv.hconcat((img6, img7))
plt.imshow(finalFrame)
```

```python
# Tugas 2
# Menampilkan image baris ke 10-100, kolom 10-100!
img8 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cropped_image = img8[10:100, 10:100]
plt.imshow(cropped_image)
```

```python
# Tugas 3
# Tampilkan image baris ke 5-30, semua kolom, channel Green saja!
img9 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img9 = img9[5:30]
img9[:, :, 0] = 0
img9[:, :, 2] = 0
plt.imshow(img9)
```

```python
# Tugas 4
# Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!
img_4 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
x1 = np.array([np.random.randint(50,100),np.random.randint(50,100),np.random.randint(250,300),np.random.randint(250,300),np.random.randint(450,500)])
y1 = np.array([np.random.randint(50,100),np.random.randint(250,300),np.random.randint(250,300),np.random.randint(350,400),np.random.randint(350,400)])
x2 = np.array([np.random.randint(150,200),np.random.randint(150,200),np.random.randint(450,500),np.random.randint(400,450),np.random.randint(600,650)])
y2 = np.array([np.random.randint(150,200),np.random.randint(400,450),np.random.randint(150,200),np.random.randint(450,500),np.random.randint(550,600)])
c1 = np.array([np.random.randint(0,200),np.random.randint(0,250),np.random.randint(0,250),np.random.randint(200,255),np.random.randint(200,255)])
c2 = np.array([np.random.randint(0,200),np.random.randint(0,250),np.random.randint(200,255),np.random.randint(200,255),np.random.randint(0,250)])
c3 = np.array([np.random.randint(0,200),np.random.randint(200,255),np.random.randint(200,255),np.random.randint(200,255),np.random.randint(0,255)])

for i in range(5):
  cv.rectangle(img_4,pt1=(x1[i],y1[i]),pt2=(x2[i],y2[i]),color=(int(c1[i]),int(c2[i]),int(c3[i])),thickness=10)

plt.imshow(img_4)
```
